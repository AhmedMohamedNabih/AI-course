```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv('/content/CAR DETAILS FROM CAR DEKHO (1).csv')
df
```

```python
sns.scatterplot(x='year',y='selling_price',data=df)
plt.show()
```

```python
sns.barplot(x='fuel',y='selling_price',data=df)
plt.title('selling price')
plt.show()
```

```python
sns.scatterplot(x='fuel',y='year',data=df)
plt.show()
```

```python
sns.violinplot(x='transmission',y='year',data=df,color='r')
```

```python

heatmap_data = df.pivot_table(index='fuel', columns='year', values='selling_price', aggfunc='mean')
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, fmt=".2f", cmap="coolwarm", linewidths=.5)
plt.title('Average Selling Price by Fuel Type and Year', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Fuel Type', fontsize=14)
plt.show()
```

```python
plt.pie(df['transmission'].value_counts(),
        labels=df['transmission'].value_counts().index,
        autopct='%0.1f%%',
        shadow=True)
plt.show()
```

```python
fig = px.scatter(df, x='transmission',y='year')
fig.show()
```

```python
fig=px.scatter(df,x='year',y='selling_price',color='transmission',
               size='year',hover_data='transmission')
fig.show()
```
